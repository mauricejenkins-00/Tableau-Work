```python
# Finally we determine the most popular start stations based on data collected from both datasets 
# Here we create new dataframes showing only the stations name and the count of rides started from its location
start_stations1 = citi_df_2021.groupby("start_station_name").count()
start_stations2 = citi_df_2020.groupby("start station name").count()
start_stations1 = start_stations1[["ride_id"]]
start_stations2 = start_stations2[["bikeid"]]
```

```python
# Cleaning up our dataframe by renaming the column holding the count values
start_stations1.rename(columns = {'ride_id':'Count'}, inplace = True)
start_stations2.rename(columns = {'bikeid':'Count'}, inplace = True)

start_stations1.head()
```

```python
# Sorting the Count column in desc order to show most popular start stations in both dataframes
start_stations1.sort_values(by='Count', ascending=False).head()
start_stations2.sort_values(by='Count', ascending=False).head()
```
